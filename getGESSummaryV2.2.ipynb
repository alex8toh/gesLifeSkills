```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('GES2016to2020 AllPolyFGPNSG-LifeSkillsV2.xlsx', '2016-2020 FGPNS - LifeSkills')
df 
```

```python
df.drop(['Msno','NMC MOE', 'School', 'Course Desc'], axis=1, inplace=True)
df 
```

```python
fg_df = df[df['Gradtype'] != 'PNS']
fg_df.drop('Gradtype', axis=1, inplace=True)
fg_df 
```

```python
def calResponseResult(df):

    all_results = df.groupby(["Survey Year", "Poly"])

    response_columns = ['Q23d_aResponse','Q23d_bResponse', 'Q23d_cResponse', 'Q23d_eResponse', \
                        'Q23d_fResponse', 'Q23d_gResponse', 'Q23d_hResponse', 'Q23d_iResponse', \
                        'Q23d_jResponse', 'Q23d_kResponse']
    column_names = ['Year', 'Poly'] 
    column_names.extend(response_columns)
    response_results = pd.DataFrame(columns=column_names)

    for yearpoly, gr in all_results:
        year = yearpoly[0]
        poly = yearpoly[1]
        response_dict = {'Year' : year, 'Poly' : poly}

        for response in response_columns:
            nW  = (gr[response] == "Well").sum()
            nVW = (gr[response] == "Very Well").sum()
            total = len(gr)
            percent = ((nW+nVW)/(total))*100
            response_dict[response] = round(percent,1)
        response_results = response_results.append(response_dict,ignore_index=True)

    return response_results
        
```

```python
def calAllPolyResponseResult(df):

    all_results = df.groupby(["Survey Year"])

    response_columns = ['Q23d_aResponse','Q23d_bResponse', 'Q23d_cResponse', 'Q23d_eResponse', \
                        'Q23d_fResponse', 'Q23d_gResponse', 'Q23d_hResponse', 'Q23d_iResponse', \
                        'Q23d_jResponse', 'Q23d_kResponse']
    
    column_names = ['Year'] 
    column_names.extend(response_columns)
    response_results = pd.DataFrame(columns=column_names)

    for year, gr in all_results:
        response_dict = {'Year' : year}

        for response in response_columns:
            nW  = (gr[response] == "Well").sum()
            nVW = (gr[response] == "Very Well").sum()
            total = len(gr)
            percent = ((nW+nVW)/(total))*100
            response_dict[response] = round(percent,1)
        response_results = response_results.append(response_dict,ignore_index=True)

    # dunno why the year chg to float, force it as int     
    response_results.Year = response_results.Year.astype(int)
    return response_results
        
```

```python
fg_allpoly_response_results = calAllPolyResponseResult(fg_df)
fg_allpoly_response_results
```

```python
response_columns = ['Q23d_aResponse','Q23d_bResponse', 'Q23d_cResponse', 'Q23d_eResponse', \
                    'Q23d_fResponse', 'Q23d_gResponse', 'Q23d_hResponse', 'Q23d_iResponse', \
                    'Q23d_jResponse']
    
for col in response_columns:
    plt.plot(fg_allpoly_response_results.Year, fg_allpoly_response_results[col])
```
